### Donner la version python de `dicho`

````python
def dicho(a,b,f,n):
    m=(a+b)/2
    err=abs(b-a)
    while err>n:
        if f(m)==0 :
            break
        elif f(a)*f(m)<0 :
            b=m
        else :
            a=m
        m=(a+b)/2
        err=abs(b-a)
    return m
    
````

````python
f=lambda x: x**3+x**2-3*x-3
x3=dicho(-2,-1.5,f,10)
x10=dicho(-2,-1.5,f,10)
x20=dicho(-2,-1.5,f,20)
print(x3,x10,x20)
#print(x3)
````

````python
def dichow (a ,b,f,eps):

    g,d = min (a,b),max(a,b)
    gau,dor = f(g),f(d)
    n = 0
    while d-g > eps :
        n += 1
        m = ( g + d )/2
        valm = f ( m )
        if gau * valm < 0 :
            d = m
            dor = valm
        else :
            g = m
            gau = valm
    
    return (g+d)/2,n
````

````python
def ptfixe(g,x0,eps,nmax):
    a=x0
    for i in range(nmax):
        x=a
        a=g(x)
        err=abs(a-x)
        print(a)
        if err<eps:
            return a,i
    return "Diverge"
````

>2) Montrer que l'équation $f(x)=0$ admet une solution unique que l'on peut obtenir en utilisant la méthode de Newton à partir >de $x0=1$.

### Solution Python

## Donner la solution entière avecun code python

````python
def newtonn(f,a,eps):
    x=a
    y=x-f(x)/fprime(x)
    while abs(x-y)>eps :
        x=y
        y=x-f(x)/fprime(x)
    return y
````

````python
import math
from sympy import *

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
sol=newtonn(f,0,0.001)
print(sol)
````
